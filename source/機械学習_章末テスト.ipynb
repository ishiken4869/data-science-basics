```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
```

```python
df_train = pd.read_csv("/content/drive/MyDrive/cancer_train.csv")
df_train.head()
```

```python
df_test = pd.read_csv("/content/drive/MyDrive/cancer_test.csv")
df_test.head()
```

```python
# モデルの宣言
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
le.fit(df_train['diagnosis'])
```

```python
# 適用
le.transform(df_train['diagnosis'])
```

```python
df_train['diagnosis'] = le.transform(df_train['diagnosis'])
```

```python
df_train.head()
```

```python
# モデルの定義
from sklearn.svm import SVC
svc = SVC()
```

```python
x_train = df_train.drop(["diagnosis"], axis=1).values
```

```python
t_train = df_train["diagnosis"].values
```

```python
x_test = df_test.values
```

```python
# モデルの学習
svc.fit(x_train, t_train)
```

```python
# モデルの検証
print('train score : ', svc.score(x_train, t_train))
```

```python
from sklearn.preprocessing import StandardScaler
std_scaler = StandardScaler()
```

```python
std_scaler.fit(x_train)
```

```python
# 標準化
x_train_std = std_scaler.transform(x_train)
x_test_std = std_scaler.transform(x_test)
```

```python
# 平均
round(x_train_std.mean())
```

```python
# 標準偏差
round(x_train_std.std())
```

```python
# モデルの定義
svc_std = SVC()
```

```python
# モデルの学習
svc_std.fit(x_train_std, t_train)
```

```python
# モデルの検証
print('train score : ', svc.score(x_train, t_train))

print('train score scaling : ', svc_std.score(x_train_std, t_train))
```

```python
y = svc_std.predict(x_test_std)
```

```python
pd.Series(y, name='pred').to_csv('submission.csv', index=None)
```
