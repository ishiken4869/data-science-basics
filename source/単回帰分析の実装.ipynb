```python
def regression_plot(filepath, colum1, colum2):
  # データの読み込み
  df = pd.read_csv(filepath)

  # データの中心化
  df_c = df - df.maen()

  # x, yを取り出す
  x = df_c[colum1]
  y = df_c[colum2]

  # 傾きを求める
  xx = x * x
  xy = x * y
  a = xy.sum() / xx.sum()

  # 散布図と直線をプロット
  plt.scatter(x, y)
  plt.plot(x, a*x, c='red')
```
